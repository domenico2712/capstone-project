```python
import os
import requests
import pandas as pd
```

```python
link_file = "https://esploradati.istat.it/SDMXWS/rest/data/41_983"
headers = {'Accept': 'application/vnd.sdmx.data+csv;version=1.0.0'}
print("Downloading data...")
```

```python
import requests

r = requests.get(link_file, headers=headers)
print(r.status_code)
```

```python
with open("../data_raw/data.csv", "wb") as f:
    f.write(r.content)

print("Data downloaded and saved to ../data_raw/data.csv")
```
